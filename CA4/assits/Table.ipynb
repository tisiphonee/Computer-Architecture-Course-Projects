```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('MainCNT.csv', dtype='str')

plt.figure(figsize=(10, 6))
plt.table(cellText=df.values,
          colLabels=df.columns,
          cellLoc='center',
          loc='center')
plt.axis('off') 

plt.title('Control signals for Main Control Unit')
plt.text(0.5, 0.95, 'Paria Pasehvarz : 810101393 , Mohsen Hassanzadeh : 810101569 ', horizontalalignment='center')

plt.savefig('Main Control Unit.png', bbox_inches='tight', dpi=3000) 
plt.show()
```
