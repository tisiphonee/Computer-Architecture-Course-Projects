```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('CU.csv', dtype='str')

plt.figure(figsize=(10, 6))
plt.table(cellText=df.values,
          colLabels=df.columns,
          cellLoc='center',
          loc='center')
plt.axis('off') 

plt.title('Control signals for CU')
plt.text(0.5, 0.95, 'Paria Pasehvarz , Mohsen Hassanzadeh', horizontalalignment='center')

plt.savefig('controller_signals_table.png', bbox_inches='tight', dpi=300) 
plt.show()
```
